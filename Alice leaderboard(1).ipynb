```python
n = int(input("Enter number of leaderboard scores: "))
scores = list(map(int, input("Enter leaderboard scores: ").split()))

# Duplicate removal
scores = sorted(list(set(scores)), reverse=True)

m = int(input("Enter number of Alice scores: "))
alice_scores = list(map(int, input("Enter Alice scores: ").split()))

for alice in alice_scores:
    rank = 1
    for i in scores:
        if alice < i:
            rank += 1
        else:
            break
    print(rank)
```
